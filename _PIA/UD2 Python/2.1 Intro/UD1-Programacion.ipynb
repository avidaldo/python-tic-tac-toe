```python
from string import ascii_uppercase


def capital_indexes(s):
    return [i for i in range(len(s)) if s[i].isupper()]

def capital_indexes_ascii(s):
    return [i for i in range(len(s)) if s[i] in ascii_uppercase]

print(capital_indexes("$ÑñÓó"))
print(capital_indexes_ascii("$ÑñÓó"))
capital_indexes("$ÑñÓó")
```
