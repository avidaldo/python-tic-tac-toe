```python
# Un array almacena un número fijo de elementos del mismo tipo.
# https://docs.python.org/3.12/library/array.html

import array 
  
arr = array.array('u', ['a', 'b', 'c'])   
for j in arr: print(j) 
```

```python
# Una lista es una colección ordenada y mutable de elementos.
# https://docs.python.org/3.12/tutorial/datastructures.html

sample_list = [10,"sid",['A','B']] 
print(sample_list)
```
