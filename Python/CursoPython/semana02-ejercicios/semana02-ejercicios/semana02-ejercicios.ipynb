<h2 style="color: #007BC4;">Ejercicio 1</h2>
<hr style="border: 0.5px solid #007BC4;">

Los tramos impositivos para la declaración de la renta en un determinado país son los siguientes:

|         Renta          | Tipo impositivo |
| :--------------------: | :-------------: |
|    Menos de 10000€     |       5%        |
| Entre 10000€ y 20000€  |       15%       |
| Entre 200000€ y 35000€ |       20%       |
| Entre 350000€ y 60000€ |       30%       |
|     Más de 60000€      |       45%       |

Escribir un programa que pregunte al usuario su renta anual y muestre por pantalla el tipo impositivo que le corresponde. 


#### Ejemplos de ejecución:

<pre>
¿Cuál es tu renta anual? 0
Error, la renta ha de ser mayor que 0.

--------------------------------------------------------------

¿Cuál es tu renta anual? 40000
Tu tipo impositivo es 30%
</pre>

```python
def tipo_impositivo(renta):
    return (None if renta<=0 else
            '5%' if renta < 10000 else
            '15%' if renta < 20000 else
            '20%' if renta < 35000 else
            '30%' if renta < 60000 else
            '45%')

renta = int(input('¿Cuál es tu renta anual? '))
if renta<1: print('Error, la renta ha de ser mayor que 0.')
else: print(f"Tu tipo impositivo es {tipo_impositivo(renta)}")
```

- Leer por teclado n números reales y almacenarlos en otra lista. Estos números representarán los elementos del segundo vector.

```python

def escalar_product(vector1, vector2, ndigits=2):
    if len(vector1) != len(vector2):
        raise ValueError('Los vectores deben tener la misma dimensión')
    return round(sum([vector1[i]*vector2[i] for i in range(len(vector1))]), ndigits)

# TEST_VECTOR_11 = [1.2, 3.5, -2.1, 6.6]
# TEST_VECTOR_12 = [3.3, -0.9, 1.8, 7.5]
# TEST_VECTOR_21 = [2.3, 1.5]
# TEST_VECTOR_22 = [2.0, -2.0]
# TEST_VECTOR_22b = [2, -2]

# print(escalar_product(TEST_VECTOR_11, TEST_VECTOR_12))
# print(escalar_product(TEST_VECTOR_21, TEST_VECTOR_22))
# print(escalar_product(TEST_VECTOR_21, TEST_VECTOR_22b))
# print(f'El producto escalar vale {escalar_product(TEST_VECTOR_21, TEST_VECTOR_22b):.2f}')

dimension = int(input('Introduce la dimensión de los vectores: '))
print('Introduce los valores del primer vector: ')
vector1 = [float(input(f"Dato {i}: ")) for i in range(dimension)]
print('Introduce los valores del segundo vector: ')
vector2 = [float(input(f"Dato {i}: ")) for i in range(dimension)]
print(f'El producto escalar vale {escalar_product(vector1, vector2):.2f}')
```

<h2 style="color: #007BC4;">Ejercicio 5</h2>
<hr style="border: 0.5px solid #007BC4;">

Realizar un programa en Python que pida al usuario la dimensión y los elementos de dos listas de números enteros (pueden tener distinta dimensión):

- Una vez leídos los datos y almacenados en las listas, se generará una nueva lista que será la intersección de ambas, es decir, contendrá aquellos elementos que se encuentren en ambas listas.
- El vector intersección se mostrará por pantalla siempre que existan elementos comunes.

NOTA: se asume que el usuario no introduce elementos repetidos dentro del mismo vector.


#### Ejemplo de ejecución:

<pre>
Introduce el numero de elementos de la lista 1 (>0):  3
Introduce los números:
Dato 1: 2
Dato 2: 4
Dato 3: 1

Introduce el numero de elementos de la lista 2 (>0):  5
Introduce los números:
Dato 1: 3
Dato 2: 3
Error, elemento repetido. Dato 2: 2
Dato 3: 9
Dato 4: 0
Dato 5: 1

La intersección de las listas es: 2 1

-----------------------------------------------------------

Introduce el numero de elementos de la lista 1 (>0):  2
Introduce los números:
Dato 0: 3
Dato 1: -2

Introduce el numero de elementos de la lista 2 (>0):  3
Introduce los números:
Dato 0: 0
Dato 1: 4
Dato 2: 5

No hay elementos comunes
</pre>


```python
def input_numbers(num_elements):
    print('Introduce los números: ')
    nums = set()
    for i in range(num_elements):
        while True:
            new = input(f"Dato {i+1}: ")  
            if new not in nums: break
            print('Error, elemento repetido. ')
        nums.add(new)
    return nums

num_elements_1 = int(input('Introduce el número de elementos de la lista 1 (>0): '))
nums_1 = input_numbers(num_elements_1)
# nums_1 = {int(input(f"Dato {i+1}: ")) for i in range(num_elements_1)} # Sin comprobar repetidos
num_elements_2 = int(input('Introduce el número de elementos de la lista 2 (>0): '))
nums_2 = input_numbers(num_elements_2)

# nums_1 = {1, 2, 3, 4, 5}
# nums_2 = {3, 4, 5, 6, 7}

print('La intersección de las listas es: ', end='')
print(*(nums_1 & nums_2))
```
