## Desplazar elementos de una lista

Implementar un programa en Python que:
- Lea por teclado un número entero (n) mayor que 0.
- Lea por teclado n números y los almacene en una lista.
- Lea por teclado un número entero (d) mayor que 0.
- Genere una nueva lista donde desplace a la derecha d posiciones (de forma circular) todos los elementos de la lista.


Ejemplo de ejecución:
```none
=============================================
Dime el numero de elementos: 8
Dime los elementos: 4 7 1 2 5 9 11 31
Dime el desplazamiento: 4
La lista resultante es: 5 9 11 31 4 7 1 2
=============================================
```

````python

def pedir_numero(msg, min):
    num = int(input(f"Introduce {msg} (>{min}): "))
    while num<=min:
        num = int(input(f"Error en el rango. Introduce {msg} (>{min}): "))    
    
    return num


def pedir_lista(dim):
    lista=[]
    for i in range(dim):
        num = int(input(f"Introduce el elemnto {i}: "))
        lista.append(num)
    return lista

def pedir_lista_linea(dim):
    cad = input("Dime los elementos: ")
    lista_strs = cad.strip().split(" ")
    if len(lista_strs) > dim:
        lista_strs = lista_strs[0:dim]
    lista_nums = list(map(int,lista_strs))

    return lista_nums


# Otra forma --> lo vemos el próximo día con alguna variante
def lee_lista_teclado(n, minimo=None, maximo=None):
    lista = []
    i = 0
    while i < n:
        for j in input().split():
            num = int(j)
            if (minimo is not None) and (maximo is not None) and (num < minimo or num > maximo):
                print(num, "es un número incorrecto, se descarta...")
            else:
                lista.append(num)
                i += 1
    return lista


def desplazar_lista(lista, desplazamiento):
    return lista[-desplazamiento:]+ lista[:-desplazamiento]

# Otra forma
def desplazar_lista_v2(orig, d):    
    n = len(orig)
    res = [0] * n
    for pos, value in enumerate(orig):
        res[(pos + d) % n] = value
    return res

def mostrar_lista(lista):
    for elemento in lista:
        print(elemento, end = " ")

def main():
    dimension = pedir_numero("Dime el número de elementos",0)
    lista = pedir_lista_linea(dimension)
    desplazamiento = pedir_numero("Dime el desplazamiento", 0)
    lista_desp = desplazar_lista(lista, desplazamiento)
    print(*lista)
````

## Tamaño palabras

Escribir un programa que:
- Pida una frase al usuario
- Cree una lista con la frase dividida en palabras
- Cree una lista con la longitud de cada palabra y la muestre por pantalla.

Ejemplo de ejecución:
```none

Palabra Longitud

Voy            3

```

````python
frase = "Hola, ¿qué tal estás?"
print("Palabra    Longitud")
print("-------------------")
for palabra in frase.split():
    print(f"{palabra:12} {len(palabra):2}")
````
